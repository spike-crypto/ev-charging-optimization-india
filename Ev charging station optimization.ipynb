```python
#Charging Price Variation Across Time Slots
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Usage Time Slot"], y=df["Charging Price (INR/kWh)"], hue=df["Usage Time Slot"], palette="magma", dodge=False)
plt.title("Charging Price Variation Across Time Slots")
plt.xlabel("Time Slot")
plt.ylabel("Price (INR/kWh)")
plt.show()
```

```python
#Correlation Matrix of EV Charging Data
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of EV Charging Data")
plt.show()
```

```python
#EV Charging Demand at Different Time Slots
plt.figure(figsize=(8,5))
sns.lineplot(x=df["Usage Time Slot"], y=df["Daily Users"], hue=df["City"], marker="o", palette="tab10")
plt.title("EV Charging Demand at Different Time Slots")
plt.xlabel("Time Slot")
plt.ylabel("Daily Users")
plt.xticks(rotation=45)
plt.legend(title="City")
plt.show()
```

```python
#Price Variation Between Fast & Slow Chargers
plt.figure(figsize=(6, 4))
sns.violinplot(x=df["Charger Type"], y=df["Charging Price (INR/kWh)"], hue=df["Charger Type"], palette="muted", legend=False)
plt.title("Price Variation Between Fast & Slow Chargers")
plt.xlabel("Charger Type")
plt.ylabel("Charging Price (INR/kWh)")
plt.show()

```
